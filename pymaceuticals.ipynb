```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_file = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
merge_file.head()
```

```python
# Checking the number of mice.
merge_file["Mouse ID"].value_counts()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_id= merge_file.loc[merge_file.duplicated(subset=['Mouse ID','Timepoint']), 'Mouse ID'].unique()
duplicate_df =pd.DataFrame(duplicate_id)
duplicate_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse=merge_file[merge_file['Mouse ID']=='g989']
dup_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df= merge_file.loc[(merge_file["Mouse ID"] !='g989')]
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df["Mouse ID"].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_groupby = clean_df.groupby("Drug Regimen")

tumor_mean = drug_groupby["Tumor Volume (mm3)"].mean()
tumor_median = drug_groupby["Tumor Volume (mm3)"].median()
tumor_variance = drug_groupby["Tumor Volume (mm3)"].var()
tumor_stddev = drug_groupby["Tumor Volume (mm3)"].std()
tumor_sem = drug_groupby["Tumor Volume (mm3)"].sem()

summary_stat_df = pd.DataFrame({"Mean Tumor Vol": tumor_mean, "Median Tumor Vol": tumor_median, 
                               "Tumor Vol Variance": tumor_variance, "Tumor Vol Std Dev": tumor_stddev,
                               "Tumor Vol SEM": tumor_sem})
summary_stat_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
aggregate = clean_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
aggregate
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tm = clean_df.groupby(['Mouse ID'])['Timepoint'].max()
final_tm_df = pd.DataFrame(final_tm)

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_df = pd.merge(clean_df, final_tm_df, on=('Mouse ID', 'Timepoint'), how='right')
final_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
treatment_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 


    # Locate the rows which contain mice on each drug and get the tumor volumes
for drug in treatments:
    drug_final=final_df.loc[final_df["Drug Regimen"]==drug, "Tumor Volume (mm3)"]
    # add subset 
    treatment_list.append(drug_final)

    
    # Determine outliers using upper and lower bounds
    quartiles= drug_final.quantile([.25,.5,.75])
    lowerq= quartiles[.25]
    upperq= quartiles[.75]
    iqr= upperq-lowerq


    lower_bound= lowerq - (1.5*iqr)
    upper_bound= upperq + (1.5*iqr)


    outlier_occupancy= drug_final.loc[(drug_final<lower_bound)|\
                                           (drug_final>upper_bound)]
    
    print(f"{drug}: {outlier_occupancy}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
green_diamond= dict(markerfacecolor="g", marker="D")

fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Vol across Regimen')
ax1.set_ylabel('Final Tumor Vol')
ax1.set_xlabel("Drug Regimen")
ax1.boxplot(treatment_list, flierprops=green_diamond)
ax1.set_xticklabels(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
cor= round(st.pearsonr(avg_df['Weight (g)'],avg_df["Tumor Volume (mm3)"])[0],4)
```

```python
cor
print(f"The correlation coefficient is {cor}")
```

```python
slope, intercept, r, p, se =st.linregress(avg_df["Weight (g)"],avg_df["Tumor Volume (mm3)"])
ln_plot= avg_df["Weight (g)"]*slope + intercept
avg_df.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)", title="Weight vs Avg Tumor Vol for Capomulin")
plt.plot(avg_df["Weight (g)"],ln_plot, color="red")
```
